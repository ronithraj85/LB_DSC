```python
import numpy
import pandas as pd
```

```python
titanic = pd.read_csv("../titanic_dataset.csv")
```

```python
titanic.head(2)
```

```python
titanic.describe(include='all')
```

```python
titanic.shape
```

```python
columns_with_nan = titanic.columns[titanic.isnull().any()]
print('Columns which have nan values are:')
for col in columns_with_nan:
    print(col)
    
```

```python
print(len(titanic) * 0.75)
print(titanic.count())
```

```python
print(titanic.count()/len(titanic)*100)
```

```python
#drop columns with more than 25% missing data
perc = len(titanic) * .75
titanic.dropna(thresh = perc, axis = 1, inplace = True)
titanic.head()
```

```python
print(titanic.count()/len(titanic)*100)
```

```python
titanic.drop(columns = ['Unnamed: 0', 'name'], inplace=True)
print('Columns after dropping independent values are:')
for col in titanic.columns:
    print(col, end = ' , ')
```

```python
# check data types of all columns
print(titanic.dtypes)
```

```python
# convert price to numeric
print(titanic.head(2))
titanic['fare'] = titanic['fare'].replace('[\$]', '', regex=True)
titanic['fare'] = pd.to_numeric(titanic['fare'])
print()
print(titanic.head(2))
```

```python
# find columns still having missing/na values and also count of missing data
missing_cols = titanic.columns[titanic.isna().any()]
print(missing_cols)
print('\nCount of missing data')
print(titanic.isnull().sum().sort_values(ascending = False).head(len(missing_cols)))
```

```python
# # fill na with mean for fare and age column column.
#for fare
print(titanic['fare'].isna().sum())
mean_fare = titanic.fare.mean()
print(mean_fare)
titanic['fare'].fillna(mean_fare, inplace = True)
# for age
print(titanic['age'].isna().sum())
mean_age = titanic.age.mean()
print(mean_age)
titanic['age'].fillna(mean_age, inplace = True)
```

```python
print('After filling the age and fare with means the missing columns are')
missing_cols = titanic.columns[titanic.isna().any()]
print(missing_cols)
print(titanic.isnull().sum().sort_values(ascending = False).head(len(missing_cols)))
```

```python
#  drop na values for embarked column.
print(titanic.shape)
print(titanic.embarked.isnull().sum())
titanic.dropna(subset = ['embarked'], inplace=True)
print(titanic.shape)
```

```python
# dump the dataframe to a csv file 'titanic_filtered.csv'.
titanic.to_csv('titanic_filtered.csv')
```

```python
print(titanic.survived.head())
# for survived column replace 0 with D and 1 with A
titanic.survived.replace([0,1],['D','A'], inplace = True)
print(titanic.survived.head())
```

```python
# find the frequency of different values in survived column
print(titanic.survived.unique())
print(titanic.survived.value_counts())
```

```python
# group by gender and survived and see the counts in each category
gender_survived_grouped = titanic.groupby(by=['gender','survived'])
for col in titanic.columns:
    print(gender_survived_grouped[col].value_counts())
```

```python
# find different pclass and no of people in each class
print(titanic['pclass'].value_counts())
```

```python
# find top 5 people with highest values of age. Count no of male and females in the top 5

a = titanic.age.sort_values(ascending = False).head()
m=f=0
for i in a.index:
    gen = titanic.iloc[i]['gender']
    if gen == 'male':
        m+=1
    else:
        f +=1
        
print(f'Number of male - {m} and female - {f}')
```

```python
titanic.sort_values('age',ascending = False).head()['gender']
```

```python
# find max age male and female who survived
print(titanic[(titanic.gender == 'male') & (titanic.survived == 'A')]['age'].max())
print(titanic[(titanic.gender == 'female') & (titanic.survived == 'A')]['age'].max())
```

```python
# get average age by gender
print(round(titanic.groupby(by='gender')['age'].mean(),2))
```

```python
# get average age by people survived vs not-survived
print(round(titanic.groupby('survived')['age'].mean(),2))
```
